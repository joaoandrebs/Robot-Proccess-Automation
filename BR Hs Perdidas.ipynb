```python
import os
import re

# Caminho da pasta de downloads
pasta_downloads = "C:/Users/lz1bkm/Downloads"

# Procurar todos os arquivos que seguem o padrão "BR21 Horas_XX"
arquivos_teste = [
    f for f in os.listdir(pasta_downloads)
    if f.startswith("BR Hs Perdidas por Empregado_") and f.endswith(".xlsx")]

for arquivo in arquivos_teste:
    caminho_antigo = os.path.join(pasta_downloads, arquivo)
    
    # Extrair o número do arquivo (_01, _02 etc.)
    match = re.search(r"_(\d{2})\.xlsx$", arquivo)
    if match:
        dia = match.group(1)
        novo_nome = f"{dia}.04.26.xlsx"  # mês e ano fixos
        caminho_novo = os.path.join(pasta_downloads, novo_nome)
        
        # Renomear o arquivo
        os.rename(caminho_antigo, caminho_novo)
        print(f"{arquivo} renomeado{novo_nome}")
    else:
        print(f"{arquivo} não corresponde ao padrão e não foi renomeado.")
```
